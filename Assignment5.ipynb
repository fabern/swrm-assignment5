##### Integration on dh and dx

$\int_{x_1}^{x_2} q dx = \int_{h1}^{h2} -k_s h dh$

the assumption of constant q and k_s:

$q \int_{x_1}^{x_2} dx = -k_s \int_{h1}^{h2} h dh$
$\Leftrightarrow$
$q(x_2-x_1) = k_s \frac{(h_2^2 - h_1^2)}{2}$ 

and finally:

$ q = \frac{k_s}{2(x_2-x_1)}(h_2^2 - h_1^2)$

We can now calculate for the situation without ($x_2-x_1=L$) and with trench ($x_2-x_1=\frac{L}{2}$):

```julia
ks = 0.002; # m/s
b = 500; # m
L = 1000; # m
h1 = 10; # m/s
h2 = 8; # m/s

h_trench = h2;

q_initial = ks/(2*L)*(h1^2 -h2^2)
q_with_trench = ks*2/(2*L)*(h1^2 -h_trench^2)

Q_initial = q_initial*b # m3/s
Q_with_trench = q_with_trench*b # m3/s
```

The trench needs to evacuate $0.036 \frac{m3}{s}$ ($36 \frac{l}{s}$).

Or $7.2e{-5} \frac{m2}{s}$ ($\frac{m3}{s}$ per meter width).
